# Data Cleaning

```python
#import the required libraries
import pandas as pd
import seaborn as sns
import re
```

```python
#drop the file 
file ='/Users/jrt/Documents/try me/Bird Strike Project 3/Bird Strikes data.xlsx'
bird_df = pd.read_excel(file, sheet_name=None)
```

```python
#displays the sheet names and their information
for sheet_name, data in bird_df.items():
    print(f'sheet name: {sheet_name}\n')
    print(data.info())
    print(f'(rows, columns): {data.shape}')
```

```python
#select the sheet for analysis 
df = bird_df['Bird Strikes']
```

```python
df.info()
```

```python
#check for the duplicate
df[df.duplicated()]
```

```python
#re-name the column names and delete the unwanted columns 
df.columns
```

```python
unwanted_columns = ['Aircraft: Type', 'Aircraft: Type', 'Wildlife: Number struck', 'Aircraft: Number of engines?', 
                    'Remains of wildlife collected?', 'Remains of wildlife sent to Smithsonian', 'Remarks',
                    'Wildlife: Size', 'Number of people injured', 'Is Aircraft Large?']
```

```python
df = df.drop(columns = unwanted_columns)
```

```python
#function to clean column names
def clean_column_names(df):
    df.columns = [re.sub(r'[^\w\s]', '', col).replace(' ','_').lower() for col in df.columns]
    return df
```

```python
df = clean_column_names(df)
```

```python
df.info()
print('data shape:', df.shape)
```

```python
re_name_columns = {'aircraft_makemodel': 'aircraft_model', 'flightdate': 'flight_date', 
                   'aircraft_airlineoperator': 'aircraft_airline', 
                   'when_phase_of_flight': 'phase_of_flight',
                   'pilot_warned_of_birds_or_wildlife': 'pilot_warned_of_wildlife',
                   'cost_total_': 'total_cost'}
```

```python
df = df.rename(columns = re_name_columns)
```

```python
df.info()
```

```python
#convert objects to string
def convert_object_to_str(df):
    for col in df.select_dtypes(include=['object']).columns:
        df[col] = df[col].astype('string')
    return df
```

```python
df = convert_object_to_str(df)
```

```python
df.info()
```

```python
total = df.isnull().sum()
percent = df.isnull().sum()/df.shape[0]*100
nulls = pd.DataFrame({'total_nulls': total,
                      'percentage': percent,
                      'data_type': df.dtypes})
```

```python
print(f'total null values, percentage and their datatypes:\n', nulls)
```

```python
#percent above the half is not considered, but we have a task to do this. so, check higher nulls
higher_nulls = ['effect_impact_to_flight', 'conditions_precipitation']
```

```python
remaining_nulls = ['airport_name', 'altitude_bin', 'aircraft_airline', 'origin_state', 'phase_of_flight',
                 'pilot_warned_of_wildlife']
```

```python
#funtion to check unique values to know the error
def for_unique_values(df, cols):
    #get the unique values in each column
    unique = {col: df[col].unique() for col in cols}
    #print unique values for each column
    for col, uni in unique.items():
        print(f'unique values in {col}: {uni}\n\n')
    return 
```

```python
for_unique_values(df, higher_nulls)
```

```python
for_unique_values(df, remaining_nulls)
```

```python
#<NA> values are 'none' where the impact of flight and conditions have no change {pandas read none as nulls/blanks}
for col in higher_nulls:
    df[col] = df[col].fillna('none')
```

```python
#removing special character in aircraft_airlines 
df['aircraft_airline'] = df['aircraft_airline'].str.replace(r'[^\w\s]','', regex = True)
```

```python
#find out remaining null values also
remaining_nulls = ['airport_name', 'altitude_bin', 'aircraft_airline', 'origin_state', 'phase_of_flight', 
                   'pilot_warned_of_wildlife']
```

```python
total = df.isnull().sum()
percent = df.isnull().sum()/df.shape[0]*100
nulls = pd.DataFrame({'total_nulls': total,
                      'percentage': percent,
                      'data_type': df.dtypes})
nulls
```

```python
#by using percentile method
upper_limit = df['wildlife_number_struck_actual'].quantile(0.99)
lower_limit = df['wildlife_number_struck_actual'].quantile(0.01)
print('upper_limit:', upper_limit)
print('lower_limit:', lower_limit)
```

```python
#finding the location of outliers
df.loc[(df['wildlife_number_struck_actual'] > upper_limit) | (df['wildlife_number_struck_actual'] < lower_limit)]
```

```python
#capping - change the outlier values to upper & lower limit vise versa
new_df = df.copy()
new_df.loc[(new_df['wildlife_number_struck_actual'] > upper_limit), 'wildlife_number_struck_actual'] = upper_limit
new_df.loc[(new_df['wildlife_number_struck_actual'] < lower_limit), 'wildlife_number_struck_actual'] = lower_limit
```

# missing values

```python
total = new_df.isnull().sum()
percent = new_df.isnull().sum()/df.shape[0]*100
nulls = pd.DataFrame({'total_nulls': total,
                      'percentage': percent,
                      'data_type': new_df.dtypes})
nulls
```

```python
string_nulls = ['airport_name', 'altitude_bin', 'aircraft_airline', 'origin_state',
                'phase_of_flight', 'pilot_warned_of_wildlife']
```

```python
for col in string_nulls:
    new_df[col] = new_df[col].fillna(new_df[col].mode()[0])
```

```python
#for numerical null values use mean or median
new_df['feet_above_ground'] = new_df['feet_above_ground'].fillna(new_df['feet_above_ground'].median())
```

```python
#for dates use forward fill or backward fill as they are in sequence
new_df['flight_date'] = new_df['flight_date'].ffill()
```
